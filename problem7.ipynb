```python
import advent_of_code_2025
lines = advent_of_code_2025.get_lines_py(7,True,1)
from functools import reduce
```

```python
tot = 0
cur = [x == 'S' for x in lines[0]]
for line in lines[1:]:
    nxt = [False] * len(cur)
    for (i,x) in enumerate(line):
        if x == '^' and cur[i]:
            tot += 1
            if i - 1 >= 0:
                nxt[i-1] = True
            if i + 1 < len(nxt):
                nxt[i+1] = True
        elif cur[i]:
            nxt[i] = True
    cur = nxt
tot
```

```python
cur = [1 if x == 'S' else 0 for x in lines[0]]
for line in lines[1:]:
    nxt = [0] * len(cur)
    for (i,x) in enumerate(line):
        if x == '^' and cur[i]:
            if i - 1 >= 0:
                nxt[i-1] += cur[i]
            if i + 1 < len(nxt):
                nxt[i+1] += cur[i]
        elif cur[i]:
            nxt[i] += cur[i]
    cur = nxt
sum(cur)
```

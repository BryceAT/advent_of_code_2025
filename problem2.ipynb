```python
import math
```

```python
tot = 0
for x in l:
    a,b = [int(s) for s in x.split('-')]
    digits = int(math.log10(a)) + 1
    div = (pow(10,digits // 2) + 1)
    top = pow(10,digits)
    odd = digits % 2 == 1
    for y in range(a, b + 1):
        if y >= top:
            odd = not odd
            top *= 10
            div = ((div -1) * 10) + 1
        if not odd and y % div == 0:
            #print(y, a, div, top)
            tot += y
tot
```

```python
pats = {1:[],
        2:[11],
        3:[111],
        4:[1111,101],
        5:[11111,],
        6:[111111,1001,10101],
        7:[1111111,],
        8:[11111111,10001,1010101],
        9:[111111111,1001001,],
        10:[1111111111,100001,101010101],
        11:[11111111111]}
tot = 0
for x in l:
    a,b = [int(s) for s in x.split('-')]
    digits = int(math.log10(a)) + 1
    top = pow(10,digits)
    for y in range(a,b+1):
        if y >= top:
            top *= 10
            digits += 1
        if any(y%p == 0 for p in pats[digits]):
            tot += y
tot
```

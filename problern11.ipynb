```python
from collections import defaultdict
```

```python
def count_paths(from_name, to_name, sink = []):
    'count # of paths from_name to_name that do not passthrough sink'
    cur = defaultdict(int,{from_name:1})
    while {k for k in cur if k != to_name}:
        nxt = defaultdict(int)
        for k,count in cur.items():
            if k in sink:
                pass
            elif k == to_name:
                nxt[k] += count
            elif k in d:
                for val in d[k]:
                    nxt[val] += count
        cur = nxt
    return cur[to_name]
count_paths('you','out')
```

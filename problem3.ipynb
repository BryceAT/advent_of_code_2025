```python
def joltage(v):
    a,b = -1,-1
    for x in v[:-1]:
        if x > a:
            a = x
            b = -1
        elif x > b:
            b = x
    if v[-1] > b:
        b = v[-1]
    return a * 10 + b
```

```python
def sub(cur: '[i32;12]', val: 'num', remaining: 'num'):
    for i in range(-min(remaining,12),0):
        if cur[i] < val:
            cur[i] = val
            for j in range(i+1,0):
                cur[j] = -1
            return cur 
    return cur
def joltage(v):
    remaining = len(v)
    cur = [-1] * 12
    for val in v:
        cur = sub(cur, val, remaining)
        remaining -= 1
    jolt = 0
    for d in cur:
        jolt = jolt * 10 + d
    return jolt
    
```

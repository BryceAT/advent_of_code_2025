```python
import advent_of_code_2025
l = advent_of_code_2025.get_lines_py(5,True,1)
```

```python
l
```

```python
intervals = []
nums = []
do_intervals = True
for line in l:
    if do_intervals:
        if '-' in line:
            a,b = line.split('-')
            intervals.append([int(a),int(b)])
        else:
            do_intervals = False
    else:
        nums.append(int(line))
```

```python
intervals
```

```python
nums
```

```python
cnt = 0
for n in nums:
    if any(a<=n<=b for a,b in intervals):
        cnt += 1
cnt
```

```python
intervals.sort()
dedup = []
for a,b in intervals:
    if len(dedup) == 0:
        dedup.append([a,b])
    elif dedup[-1][1] + 1 >= a:
        dedup[-1][1] = max(dedup[-1][1],b)
    else:
        dedup.append([a,b])
```

```python
sum(b-a + 1 for a,b in dedup)
```

```python
import advent_of_code_2025
lines = advent_of_code_2025.get_lines_py(6,True,1)
from functools import reduce
```

```python
lines
```

```python
m = [l.split() for l in lines]
for i in range(len(m) - 1):
    m[i] = [int(x) for x in m[i]]
t = []
for c in range(len(m[0])):
    cur = []
    for i in range(len(m)):
        cur.append(m[i][c])
    t.append(cur)
t
```

```python
sum(reduce(lambda x,y: x*y,row[:-1],1) if row[-1] == '*' else reduce(lambda x,y: x+y,row[:-1],0) for row in t)
```

```python
t = []
chunk = []
for c in range(len(lines[0])):
    if c < len(lines[-1]) and lines[-1][c] in ['+','*']:
        if chunk:
            t.append(chunk)
        chunk = [lines[-1][c]]
    cur = 0
    for r in range(len(lines)-1):
        if '0' <= lines[r][c] <= '9':
            cur = cur * 10 + int(lines[r][c])
    if cur > 0:
        chunk.append(cur)
t.append(chunk)
t
```

```python
sum(reduce(lambda x,y: x*y,row[1:],1) if row[0] == '*' else reduce(lambda x,y: x+y, row[1:],0) for row in t )
```

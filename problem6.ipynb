```python
m = [l.split() for l in lines]
for i in range(len(m) - 1):
    m[i] = [int(x) for x in m[i]]
t = []
for c in range(len(m[0])):
    cur = []
    for i in range(len(m)):
        cur.append(m[i][c])
    t.append(cur)
```

```python
tot = 0
t = []
for c in range(len(lines[0])-1,-1,-1):
    cur = 0
    is_empty = True
    for r in range(len(lines)):
        match lines[r][c] if len(lines[r]) > c else ' ':
            case '*':
                t.append(cur)
                is_empty = True
                tot += reduce(lambda x,y: x*y, t,1)
                t = []
            case '+':
                t.append(cur)
                is_empty = True
                tot += reduce(lambda x,y: x+y,t,0)
                t = []
            case x if '0'<=x<='9':
                cur = cur * 10 + int(x)
                is_empty = False
            case _:
                pass
    if not is_empty:
        t.append(cur)
tot
```

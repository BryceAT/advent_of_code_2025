```python
g = advent_of_code_2025.CharGrid(grid)
n = len(grid)
m = len(grid[0])
```

```python
len([(x,y) for x in range(n) for y in range(m) if g.get(x,y) == '@' and g.count_neighbors(x,y,'@') < 4])
```

```python
tot = 0
while (can_be_removed := len([g.set(x,y,'.') for x in range(n) for y in range(m) if g.get(x,y) == '@' and g.count_neighbors(x,y,'@') < 4])) > 0:
    tot += can_be_removed
tot
```

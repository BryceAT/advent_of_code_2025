```python
shapes = {}
regions = []
cur = 0
shape = []
for line in lines:
    if len(line) == 2:
        cur = int(line[0])
    elif line == '' and shape:
        shapes[cur] = [shape]
        shape = []
    elif len(line) > 3:
        s = [int(x) for x in line.split(':')[0].split('x')]
        cnts = [int(x) for x in line.split(':')[1].split()]
        regions.append((s,cnts))
    else:
        shape.append([x == '#' for x in line])
def rotate(shape):
    out = [[x for x in row] for row in reversed(shape)]
    for i in range(len(shape)):
        for j in range(i):
            out[i][j],out[j][i] = out[j][i],out[i][j]
    return out
def mk_dihedral(i):
    for _ in range(3):
        shapes[i].append(rotate(shapes[i][-1]))
    shape = list(reversed(shapes[i][0]))
    shapes[i].append(shape)
    for _ in range(3):
        shapes[i].append(rotate(shapes[i][-1]))
for i in shapes:
    mk_dihedral(i)
def print_shape(shape):
    for row in shape:
        print(''.join(['X' if x else '.' for x in row]))
```

```python
# counting pips is can only rule out regions that are too small, but in this case it seems to be sufficent as well
shapes_cnts = {i:sum(1 if x else 0 for row in shape_types[0] for x in row) for i,shape_types in shapes.items()}
def fit_pips(region):
    return sum(shapes_cnts[i]*cnt for i,cnt in enumerate(region[1])) <= region[0][0] * region[0][1]

sum(1 for region in regions if fit_pips(region))
```

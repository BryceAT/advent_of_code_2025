```python
import advent_of_code_2025
points = advent_of_code_2025.get_csv_muliline_py(8,True,1)
```

```python
points
```

```python
uf = advent_of_code_2025.UnionFind(len(points))
```

```python
def dist(x,y):
    return (x[0]-y[0])**2 + (x[1]-y[1])**2 + (x[2]-y[2])**2
pairs = [(dist(points[i],points[j]),i,j) for i in range(len(points)) for j in range(i + 1, len(points))]
pairs.sort()
```

```python
joins_to_make = 10
for _,i,j in pairs[:joins_to_make]:
    uf.union(i,j)
group_sizes = uf.get_sizes()
group_sizes.sort()
#print(group_sizes)
group_sizes[-1] * group_sizes[-2] * group_sizes[-3]
```

```python
for _,i,j in pairs[joins_to_make:]:
    uf.union(i,j)
    if uf.get_max_size() == len(points):
        break 
(i,j) = uf.get_last_union()
points[i][0] * points[j][0]
```

```python
def dist(x,y):
    return (x[0]-y[0])**2 + (x[1]-y[1])**2 + (x[2]-y[2])**2
pairs = [(dist(points[i],points[j]),i,j) for i in range(len(points)) for j in range(i + 1, len(points))]
pairs.sort()
```

```python
joins_to_make = 10
parents = list(range(len(points)))
sizes = [1] * len(points)
def find(x):
    if parents[x] != x:
        parents[x] = find(parents[x])
    return parents[x]
def union(x,y):
    x = find(x)
    y = find(y)
    if x != y:
        x,y = min(x,y),max(x,y)
        sizes[y] += sizes[x]
        sizes[x] = 0
        parents[x] = y
for _,i,j in pairs[:joins_to_make]:
    union(i,j)
group_sizes = []
for i in range(len(points)):
    if find(i) == i:
        group_sizes.append(sizes[i])
group_sizes.sort()
group_sizes
```
